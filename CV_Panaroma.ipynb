# **Importing Libraries**

```python
# import the necessary packages
from imutils import paths
import numpy as np
import argparse
import imutils
import cv2
from google.colab.patches import cv2_imshow
```

# **Panorama using Image Stitching**

```python
img1=cv2.imread("1_pn.jpg")
img2=cv2.imread("2_pn.jpg")
img3=cv2.imread("3_pn.jpg")
img4=cv2.imread("4_pn.jpg")
img5=cv2.imread("5_pn.jpg")

print(img1.shape)
print(img2.shape)
print(img3.shape)
print(img4.shape)
print(img5.shape)

imgg1=cv2.resize(img1,(500,300))
imgg2=cv2.resize(img2,(500,300))
imgg3=cv2.resize(img3,(500,300))
imgg4=cv2.resize(img4,(500,300))
imgg5=cv2.resize(img5,(500,300))
#print(imgg1.shape)
#cv2_imshow(imgg1)
```

```python
images=[]
images.append(imgg1)
images.append(imgg2)
images.append(imgg3)
images.append(imgg4)
images.append(imgg5)
for i in images:
  cv2_imshow(i)
```

```python
stitcher=cv2.Stitcher.create(cv2.Stitcher_PANORAMA)
stitcher.setPanoConfidenceThresh(0.3) 
images1 = []
images1.append(images[0])
images1.append(images[1])
(status1,out1)=stitcher.stitch(images1)
cv2_imshow(out1)
```

```python
stitcher=cv2.Stitcher.create(cv2.Stitcher_PANORAMA)
stitcher.setPanoConfidenceThresh(0.3)

images2 = []
images2.append(images[1])
images2.append(images[2])
(status2,out2)=stitcher.stitch(images2)
cv2_imshow(out2)
```

```python
stitcher=cv2.Stitcher.create(cv2.Stitcher_PANORAMA)
stitcher.setPanoConfidenceThresh(0) 

images3 = []
images3.append(images[2])
images3.append(images[3])
(status3,out3)=stitcher.stitch(images3)
cv2_imshow(out3)
```

```python
stitcher=cv2.Stitcher.create(cv2.Stitcher_PANORAMA)
stitcher.setPanoConfidenceThresh(0.5)

images4 = []
images4.append(images[3])
images4.append(images[4])
(status4,out4)=stitcher.stitch(images4)
cv2_imshow(out4)
```

Combine all the stitches to get the final stitched Panorama using 5 images

```python
stitched1=cv2.resize(out1,(500,300))
stitched2=cv2.resize(out2,(500,300))
stitched3=cv2.resize(out3,(500,300))
stitched4=cv2.resize(out4,(500,300))

combined_stitched = []
combined_stitched.append(stitched1)
combined_stitched.append(stitched2)
combined_stitched.append(stitched3)
combined_stitched.append(stitched4)

(status5,final_out)=stitcher.stitch(combined_stitched)
cv2_imshow(final_out)
```
